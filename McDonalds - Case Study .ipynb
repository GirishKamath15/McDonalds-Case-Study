```python
df=data.iloc[:,11:13]
```

```python
full_data=pd.concat([cat_dummy,df],axis=1)
```

```python
final_data=pd.concat([cat_dummy,df],axis=1)
```

```python
final_data.head()
```

```python
final_data.corr()
```

```python
plt.figure(figsize=(10,10))
sns.heatmap(final_data.corr(),annot = True, cmap = 'Spectral')
```

```python
from sklearn.preprocessing import StandardScaler
scalar= StandardScaler()

scaled_data = scalar.fit_transform(final_data)
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=11)
principal_components = pca.fit_transform(scaled_data)
```

```python
pc=pd.DataFrame(principal_components)
```

```python
pc.head()
```

```python
pc.rename(columns={0:"PCA1",1:"PCA2",2:"PCA3",3:"PCA4",4:"PCA5",5:"PCA6",6:"PCA7",7:"PCA8",8:"PCA9",9:"PCA10",10:"PCA11"}, inplace = True)
```

```python
pc.head()
```

```python
from sklearn.cluster import KMeans
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters = i)
    kmeans.fit(pc)
    wcss.append(kmeans.inertia_)
```

```python
plt.figure(figsize=(5,5))
plt.plot(range(1,11),wcss)
plt.xlabel("number of k (cluster) value")
plt.ylabel("wcss")
plt.show()
```

```python
KMeans_Model=KMeans(2)
KMeans_Model.fit_predict(pc)
```

```python
KMeans_Model.labels_
```

```python
labels=pd.DataFrame(KMeans_Model.labels_)
```

```python
new_pc=pd.concat([pc,labels], axis=1)
```

```python
new_pc.rename(columns={0:"labels"}, inplace = True)
```

```python
new_pc.head()
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="PCA1",y="PCA3",data=new_pc, hue="labels")
plt.title("Clustering using K-Means")
plt.plot()
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram
merg = linkage(new_pc,method="ward")
plt.figure(figsize=(25,10))
dendrogram(merg,leaf_rotation = 90)
plt.xlabel("data points")
plt.ylabel("wcss")
plt.show()
```

```python
full_data.head()
```

```python
full_data["labels"]=new_pc["labels"]
```

```python
full_data.head()
```

```python
from sklearn.model_selection import train_test_split

X = full_data.drop(['labels'],axis=1)
Y= full_data['labels']
x_train, x_test, y_train, y_test =train_test_split(X, Y, test_size=0.3,random_state =123)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(x_train,y_train)
```

```python
dt_y_pred=dt.predict(x_test)
```

```python
from sklearn.metrics import classification_report
from sklearn.metrics import confusion_matrix
```

```python
print(classification_report(y_test,dt_y_pred))
```

```python
matrix=confusion_matrix(y_test,dt_y_pred)
print(matrix)
```
